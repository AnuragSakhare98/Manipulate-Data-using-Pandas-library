```python
# This method in pandas is used to remove duplicate rows from a DataFrame.
# By default, it considers all columns to identify duplicate rows, but it can also be configured to consider specific columns.
# The method returns a DataFrame with the duplicate rows removed.
DataSet = DataSet.drop_duplicates()
print(DataSet)
```

```python
# This method in pandas is used to detect missing values in a DataFrame.
# It will return the elements in boolean values i.e. True and False where True means that element is (NaN) or none and False means that element is not null.
# This method is useful for identifying and handling missing data in your dataset.
DataSet.isnull()
```

```python
# From the above syntax it is hard to find how many elements in particular column are null in the DataSet, so there is an another syntax
# below where we can count the no. of missing values in particular column
```

```python
DataSet.isnull().sum() # To count the number of missing values in each column
```

```python
DataSet.notnull() # Returns a DataFrame with boolean values, where True indicates non-missing values and False Indicates null values
```

```python
# This method is used to count the number of non-null (non-missing) entries in each column of a DataFrame.
# This is useful for quickly getting an overview of the completeness of data.
DataSet.notnull().sum()
```

```python
# "DataSet.isnull()": This method returns a DataFrame of the same shape as DataSet, with boolean values indicating,
# where the missing values (NaNs) are (True for missing, False for non-missing).


# "sum().sum()": The first sum() calculates the sum of True values (which are treated as 1) for each column,
# giving a Series where each element represents the number of missing values in a column.
# The second sum() then calculates the sum of this Series, giving the total number of missing values in the entire DataFrame.

DataSet.isnull().sum().sum() # This method is used to calculate the total number of missing values in an entire DataFrame.
```

```python
DataSet1 = DataSet.fillna(value=0) # fillna() is used to fill missing (NaN) values in the DataFrame or Series.
                                   # (value=0) is a parameter of the fillna() method. It specifies the value
                                   # that will be used to fill in the missing values. In this case,we are filling missing values with 0.

print(DataSet1)
```

```python
#Detect the outliers using IQR
DataSet1.columns
```

```python
# This syntax in pandas is used to drop a column named 'Observation' from the DataFrame DataSet1.
# This operation removes the specified column from the DataFrame.
# The inplace=True parameter ensures that the operation is performed directly on the original DataFrame DataSet1 without the need for reassignment.

DataSet1.drop(['Observation'], axis = 1, inplace = True)
```

```python
DataSet1.columns
```

```python
#'DataSet1.quantile(0.25)': This method call calculates the first quartile (Q1) for each column in the DataFrame DataSet1.
# The parameter 0.25 specifies that the first quartile should be calculated, which corresponds to the 25th percentile.

Q_1 = DataSet1.quantile(0.25)

# DataSet1.quantile(0.75): Similarly, this method call calculates the third quartile (Q3) for each column in the DataFrame DataSet1.
# The parameter 0.75 specifies that the third quartile should be calculated, which corresponds to the 75th percentile.
Q_3 = DataSet1.quantile(0.75)

# The IQR is a measure of statistical dispersion and is often used in descriptive statistics to identify the spread of the middle 50% of the dataset.
# Calculate IQR
IQR = Q_3 - Q_1
print("IQR is:\n",IQR)
```

```python
# This code performs outlier detection and removal in the DataFrame 'DataSet1' based on the Interquartile Range (IQR) method
DataSet1 = DataSet1 [ ~ ((DataSet1 < (Q_1 - 1.5 * IQR)) | (DataSet1 > (Q_3 + 1.5* IQR))).any (axis = 1)]
print(DataSet1)
```
